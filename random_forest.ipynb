```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import plotly.figure_factory as ff
import seaborn as sns
import sklearn
```

```python
df= pd.read_csv('heart.csv')
```

```python
object_columns = df.select_dtypes(['object']).columns
df[object_columns] = df[object_columns].astype('category')
```

```python
df_copy=df.copy()
```

```python
data=df_copy
target = pd.DataFrame(data.pop('HeartDisease'))
data
```

```python
data= pd.get_dummies(data, drop_first=True)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data, target)
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators=100)
classifier.fit(X_train, y_train)
```

```python

from sklearn.metrics import accuracy_score

y_pred = classifier.predict(X_test)

accuracy_score(y_test, y_pred)
```

```python
classifier.feature_importances_
```

```python
import plotly.express as px
features = pd.DataFrame(data={'feature': data.columns, 'feature_importance': classifier.feature_importances_})

px.bar(features, x='feature', y='feature_importance', width=700, height=400)
```

```python
new_df = data.drop(['RestingECG_ST', 'RestingECG_Normal', 'ChestPainType_TA', 'ChestPainType_NAP', 'ChestPainType_ATA'], axis=1)
new_df
```

```python
X_train_new, X_test_new, y_train_new, y_test_new = train_test_split(new_df, target)
```

```python
new_classifier = RandomForestClassifier(n_estimators=100)
new_classifier.fit(X_train_new, y_train_new)
```

```python
y_pred_new = new_classifier.predict(X_test_new)

accuracy_score(y_test_new, y_pred_new)
```

```python
new_classifier.feature_importances_
```

```python
import plotly.express as px
features = pd.DataFrame(data={'feature': new_df.columns, 'feature_importance': new_classifier.feature_importances_})

px.bar(features, x='feature', y='feature_importance', width=700, height=400)
```
